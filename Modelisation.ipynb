```python
import fonctions as f
```

```python
df = f.pd.read_csv("data.csv")
df_cleaned = df.dropna(axis=1, how='all')
df_cleaned.head()
```

```python
df_cleaned.head()
```

```python
# Fonction pour convertir les chaînes avec des virgules en float
def convertir_virgule_en_float(serie):
    return f.pd.to_numeric(serie.str.replace(",", ".", regex=False), errors="coerce")

# Liste des colonnes à traiter
colonnes_a_corriger = ['RR', 'TN', 'TX', 'TM', 'TMNX', 'TAMPLI', 'TNTXM', 'FFM', 'FXI', 'FXY', 'FXI3S', 'TSVM']

# Application de la fonction sur les colonnes concernées
df_cleaned[colonnes_a_corriger] = df_cleaned[colonnes_a_corriger].apply(convertir_virgule_en_float)
```

```python
df_cleaned.head()
```

```python
df_cleaned["TX"].dtype 
```

```python
#Création de la variable "event":

# Définir les seuils arbitraires!
seuil_chaleur = 40  
seuil_froid = -5  
seuil_inondation = 50
seuil_tempete = 30


# Créer une variable "event"
df_cleaned["event"] = ((df_cleaned["TX"] > seuil_chaleur) | 
                        (df_cleaned["TN"] < seuil_froid) |
                        (df_cleaned["RR"] > seuil_inondation) |
                        (df_cleaned["FXI"] > seuil_tempete)).astype(int)

# Afficher le DataFrame
df_cleaned.head()
```
